Firstly we have merged the datasets t1 and uid, keeping the user_id as a key.

We have now merged the datasets t1 and ap with keeping project id as a key. As a result we have the boarding and creation time side by side.

```python
p2=t1.merge(ap[['project_id','created_at']],how='left',left_on='project_id', right_on='project_id')
p2
```

By importing the reuired libraries we have calculated the development phase time by subtracting the boarding time to the project created time and dividing the value by 60 to get the final time in seconds.

```python
import datetime
from datetime import date
from datetime import datetime

p2['created_at'] = pd.to_datetime(p2['created_at'])
p2['boarded_at'] = pd.to_datetime(p2['boarded_at'])
p2.eval("development_phase=abs(created_at-boarded_at)", engine='python', inplace=True)
p2['development_phase'] = (p2['development_phase'].dt.total_seconds())//60

p2
```
